```python
import numpy as np
quest = np.loadtxt("quest.csv", delimiter=",", dtype=np.int64)
print(quest)
```

```python
# 각 행들의 합을 구함
a = quest.sum(axis=1)
print(a)
# 각 열들의 합을 구함
b = quest.sum(axis=0)
print(b)
```

```python
# 전체 성분(데이터)의 평균
quest.mean()
```

```python
# 전체 성분(데이터) 중 최솟값
quest.min()
```

```python
# 전체 성분(데이터) 중 최댓값
quest.max()
```

```python
# 전체 성분(데이터) 의 누적합
quest.cumsum()
```

```python
# 첫 번째 성분(데이터) 부터 각 데이터까지의 누적합
quest.cumprod()
```

```python
# 각 성분(데이터)의 크기 비교
quest > 5
```

```python
# 인덱싱을 활용해 5보다 큰 수만 가져오기
quest[quest > 5]
```

```python
# 5보다 큰 숫자를 7로 바꾸기
quest[quest > 5] = 7
quest
```
