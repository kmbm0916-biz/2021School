**일주일간의 유동인구수를 리스트에 저장 후 출력**

 - 유동 인구 데이터를 컴퓨터로 저장한 후 출력하기

```python
h = [242, 256, 237, 223, 263, 81, 46]
print(h)
```

 - 유동인구수 데이터의 총합과 평균을 리스트에 추가하기

```python
h = [242, 256, 237, 223, 263, 81, 46]
h.append(sum(h))
h.append(h[(len(h)-1)]/(len(h)-1))
print(h)
```

 - 그래프로 나타내기

```python
# 한글 폰트깨짐 수정용
# !sudo apt-get install -y fonts-nanum
# !sudo fc-cache -fv
# !rm ~/.cache/matplotlib -rf
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 

week = ["월", "화", "수", "목", "금", "토", "일"]
h_plotset = h[:7]

plt.title("일주일간 유동 인구수 데이터")

plt.plot(week, h_plotset, "o-")

plt.xlabel("요일")
plt.ylabel("유동 인구수")

plt.show()
```
