# 4 리스트
**4-1 리스트 선언 및 출력**

```
subway = ['연신내역', '구파발역', '불광역'] # 변수
print(subway)

# 자료형 출력
print(type(subway))
```



---



**4-2 리스트의 함수 활용**

```
# 리스트 내부의 index(N)의 N에 해당하는 인덱스 값 출력
print(subway.index('구파발역'))

# 리스트 내부에 append(N)의 N에 해당하는 값을 추가
subway.append('3호선')
print(subway)

# 리스트 내부에 insert(N, M)의 N에 해당하는 값을 인덱스 위치로 하여, M에 해당하는 값을 추가
subway.insert(0, '지하철역')
print(subway)

# 리스트 내부의 
print(subway.pop())
print(subway)

# 리스트 내부의 count(N)의 N에 해당하는 값이 인덱스 내부에 몇 개나 있는지 출력
print(subway.count('불광역'))

# 리스트 내부의 값들을 정렬한다
subway.sort()
print(subway)

# 리스트 내부의 값을 반대로 출력한다.
subway.reverse()
print(subway)
```
