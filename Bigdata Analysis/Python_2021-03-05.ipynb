**2-2 변수 과제**

- 변수명 : station
- 변수값 : 구파발
- 출력문장 : "구파발행 열차가 들어오고 있습니다."
- 변수값 변경 : 인천, 안산

```
station = "구파발"
print(station + "행 열차가 들어오고 있습니다.")

station = "인천"
print(station + "행 열차가 들어오고 있습니다.")

station = "안산"
print(station + "행 열차가 들어오고 있습니다.")
```
