# 기본 그래프 작성 단계
1. 라이브러리 불러오기
2. plot() 함수에 데이터 입력하기 : plt.plot([x축 데이터], [y축 데이터])
3. 그래프 보여주기 : plt.show()

```python
import matplotlib.pyplot as plt
# 별명(alias)사용하기, 별칭
# 별명 사용 이유 : 모듈 사용할 때 짧은 타이핑을 위해 일반적으로 모듈명이 같음.
import matplotlib.pyplot as plt
```

```python
# 기본 그래프 그리기
# plt.plot([10, 20, 30, 40])
plt.plot([40, 30, 20, 10])
plt.show()

# x축, y축 구분하기
# x축 : plot()내 리스트의 인덱스, 자동으로 1씩 증가
# y축 : plot()내 리스트의 값
```

```python
plt.plot([10, 20, 30, 40], [12, 43, 64, 22])
plt.show()
```

# 그래프에 옵션 추가하기
 - 그래프에 제목 넣기 : plt.title()

```python
import matplotlib.pyplot as plt

plt.title('Graph Title')

plt.plot([10, 20, 30, 40])
plt.show()
```

 - 그래프에 x축, y축 이름 넣기 plt.xlabel(), plt.ylabel()

```python
import matplotlib.pyplot as plt

plt.title("Graph Title")
plt.plot([10, 20, 30, 40])

plt.xlabel("X-Label : xxx")
plt.ylabel("Y-Label : yyy")

plt.show()
```

 - 축 범위 설정하기
> - plt.axis([xmin, xmax, ymin, ymax])
> - 값을 설정하지 않으면 자동으로 지정

```python
import matplotlib.pyplot as plt

plt.title("Graph Title")
plt.plot([10, 20, 30, 40], [3, 8, 20, 70])

plt.xlabel("X-Label : xxx")
plt.ylabel("Y-Label : yyy")

plt.axis([0, 50, 0, 75])

plt.show()
```

 - 그래프에 범례 넣기 : plt.legend()

```python
import matplotlib.pyplot as plt

plt.title("Graph Title")
plt.plot([10, 20, 30, 40], label="ASC")
plt.plot([40, 30, 20, 10], label="DESC")

plt.legend()

plt.xlabel("X-Label : xxx")
plt.ylabel("Y-Label : yyy")

plt.show()
```

 - 그래프 색상 바꾸기

```python
import matplotlib.pyplot as plt

plt.title("Graph Title")
plt.plot([10, 20, 30, 40], label="ASC", color="red")
plt.plot([40, 30, 20, 10], "blue", label="DESC")

plt.legend()

plt.xlabel("X-Label : xxx")
plt.ylabel("Y-Label : yyy")

plt.show()
```

 - 그래프 선 모양 바꾸기

```python
import matplotlib.pyplot as plt

plt.title("Graph Title")
plt.plot([10, 20, 30, 40], label="ASC", color="red", linestyle="--")
plt.plot([40, 30, 20, 10], label="DESC", color="blue", ls=":")

plt.legend()

plt.xlabel("X-Label : xxx")
plt.ylabel("Y-Label : yyy")

plt.show()
```

 - 마커 모양 바꾸기

```python
import matplotlib.pyplot as plt

plt.title("Graph Title")
plt.plot([10, 20, 30, 40], "rx", label="ASC")
plt.plot([40, 30, 20, 10], "g+", label="DESC")

plt.legend()

plt.xlabel("X-Label : xxx")
plt.ylabel("Y-Label : yyy")

plt.show()
```

 - grid(그리드) 설정하기

```python
import matplotlib.pyplot as plt

plt.title("Graph Title")
plt.plot([10, 20, 30, 40], [40, 30, 20, 10], "bo")

plt.xlabel("X-Label : xxx")
plt.ylabel("Y-Label : yyy")

plt.grid(True)

plt.show()
```

- 기본적인 산점도 시각화

```python
import matplotlib.pyplot as plt

xpoints = ([2, 3, 5, 7])
ypoints = ([15, 18, 25, 35])

plt.scatter(xpoints, ypoints, c = "pink")
plt.show()
```

```python
import matplotlib.pyplot as plt

xpoints = ([2, 3, 5, 7])
ypoints = ([15, 18, 25, 35])
zpoints = ([60, 200, 700, 250])

plt.scatter(xpoints, ypoints, zpoints)
plt.show()
```

 - 막대그래프

```python
import matplotlib.pyplot as plt

xpoints = (["2019", "2020", "2021", "2022"])
ypoints = ([15, 18, 25, 35])

# plt.bar(xpoints, ypoints)
plt.bar(xpoints, ypoints, align = "edge")
plt.show()
```

 - 파이차트 그리기

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ["apple", "banana", "melon", "grape"]

# plt.pie(ratio, labels = labels)
# 시계방향으로 그림
plt.pie(ratio, labels = labels, autopct = "%.1f%%", counterclock=False)
plt.show()
```

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ["apple", "banana", "melon", "grape"]
explode = [0, 0.10, 0, 0.10]
# 반지름의 10% 벗어나도록 설정 : 0.10

# plt.pie(ratio, labels = labels)
# 시계방향으로 그림
plt.pie(ratio, labels = labels, autopct = "%.1f%%", explode = explode, startangle = 90, counterclock = False)
plt.show()
```
