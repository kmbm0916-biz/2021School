# 문제 : 자판기 작동 과정 만들기
- 커피한잔의 가격은 300원이다.
- 커피는 4잔을 판매하면 판매 중지 문구를 띄문ㅇ=ㅏ,
- 출력형태 : 돈을 넣어주세요 : 500
- 거스름돈 200을 주고 커피를 줍니다.
- 돈을 넣어주세요 : 200
- 돈을 다시 돌려주고 커피를 주지 않습니다.
- 남은 커피의 양은 3개입니다.
- 커피가 다 떨어졌습니다. 판매를 중지합니다.

```python
co_p = 300

cnt = 4

while cnt > 0:
    co_i = int(input("돈을 넣어주세요 : "))
    if co_i >= co_p:
        print("거스름돈 {}을 주고 커피를 줍니다.".format(co_i - co_p))
        cnt -= 1
        print("남은 커피의 양은 {}개입니다.".format(cnt))
    elif co_i < co_p:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
print("커피가 다 떨어졌습니다. 판매를 중지합니다.")

# 선생님 코드
# cof = 4
# while True:
#     money = int(input("돈을 넣어주세요 : "))
#     if money == 300:
#         print("커피를 줍니다.")
#         cof -= 1
#     elif money > 300:
#         print("거스름돈 {}을 주고 커피를 줍니다.".format(money - 300))
#         cof -= 1
#     else:
#         print("돈을 돌려주고, 커피를 주지 않습니다.")
#         print("남은 커피의 양은 {}개 입니다.".format(cof))

#     if not cof:
#         print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
#         break
```

# list comprehension
- 리스트를 쉽게, 짧게 한 줄로 만들 수 있는 파이썬의 문법
- 기존 리스트형을 사용하여 간단하게 새로운 리스트를 만드는 기법
- [(변수를 활용한 값) for (사용할 변수 이름) in (순회할 수 있는 값)]

```python
result = []

for i in range(10):
    result.append(i)
print(result)

# Same way to use list comprehension
# result = [i for i in range(10)]
# print(result)
```

#다음과 같이 출력되도록 for문으로 작성하고 List Comprehension으로 나타내시오.
- a = [1, 2, 3, 4, 5, 6, 7, 8, 9]를 이용한다.
- 짝수만 빼서 2배한 값으로 리스트를 만들어라
- 출력리스트 : [4, 8, 12, 16]

```python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# result = []
# for i in a:
#     if i % 2 == 0:
#         result.append(i * 2)
# print(result)

result = []
result = [i * 2 for i in a if i % 2 == 0]
print(result)
```
