**9-2 과제**
 - 은행 계좌 만들기
*   계좌 만들기
> 1. argument : 필요없음
> 2. 출력 : 새로운 계좌 생성
*   입금
> 1. argument : 잔액, 입금액
> 2. 출력 : 입금 완료
> 3. return : 잔액 + 입금액
*   입금
> 1. argument : 잔액, 출금액
> 2. 잔액이 출금액보다 크면
>>출력 : 출금 완료
>>return : 잔액 - 출금액
> 3. 잔액이 작으면
>>출력 : 출금불가
>>return : 잔액

```python
balance = 0
money = 0

# W1 계좌생성
def open_account():
    print("Create New Account")

# open_account()

# W2 입금
def deposit(balance, money):
    print("Deposit Completed.")
    return balance + money
# W3 출금
def withdraw(balance, money):
    if balance >= money:
        print("Withdraw Completed.")
        return balance - money
    else:
        print("Withdraw Incompleted.")
        return balance
# W4 잔액확인
def open_balance():
    print(balance)
# W5 저녁 출금
def withdraw_night(balance, money):
    commission = 1000
    return commission, balance - money - commission

# balance = 0
# money = input()
# balance = deposit(balance, int(money))
# money = input()
# balance = withdraw(balance, int(money))
# open_balance()

index = 1
while index == 1:
    str_c = input()
    if str_c == "입금":
        money = input()
        balance = deposit(balance, int(money))
    elif str_c == "출금":
        money = input()
        balance = withdraw(balance, int(money))
    elif str_c == "잔액확인":
        open_balance()
    elif str_c == "저녁출금":
        money = input()
        commission, balance = withdraw_night(balance, int(money))
        print("Fee : {}, balance : {}".format(commission, balance)) 
    elif str_c == "종료":
        index = index - 1

# commission, balance = withdraw_night(balance, 1000)
# print("수수료 : {}, 잔액 : {}".format(commission, balance))
```

**9-4 함수 가변인자**

```python
def profile(name, age, lang1, lang2, lang3, lang4, lang5):
    print("Name : {}, Age : {}, lang : ".format(name, age), end="")
    print(lang1, lang2, lang3, lang4, lang5)

profile("유지석", 45, "Python", "Java", "C", "C++", "C#")
profile("kkkkkk", 17, "Java", "C", "", "", "")
```

```python
def profile(name, age, *language):
    print("Name : {}, Age : {}, lang : ".format(name, age), end="")
    for lang in language:
        print(lang, end="/")
    print()

profile("유지석", 45, "Python", "Java", "C", "C++", "C#")
profile("kkkkkk", 17, "Java", "C")
profile("유재석", 45, "Python", "Java", "C", "C++", "C#", "Python", "Java", "C", "C++", "C#")
```
