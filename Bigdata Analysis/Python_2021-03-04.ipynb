**1-3 간단한 비교**


```
print(10 > 8)
```

```
print(20 > 30)
```

```
print(True)
print(False)
```

```
print(not(5 > 10))
# not = True -> False, False -> True
```

```
print(1 != 3)
```

```
print((3 > 0) and (3 > 5))
# and = &&
```

```
print((3 > 0) or (3 > 5))
# or = ||
```

```
print(True - True)
print(False + False)
print(True * True)
print(False * False)
print(True - False)
print(True + False)
# Boolean끼리의 연산도 가능
```
