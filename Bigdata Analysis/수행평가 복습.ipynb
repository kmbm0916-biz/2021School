> 사이트별로 비밀번호를 만들어 주는 프로그램을 작성하라.
 - "http://"부분은 제외 --> naver.com
 - 처음 만나는 점(.)이후 부분은 제외 --> naver
 - 남은 글자 중 처음 세자리 + 글자 개수 + 글자 내 'e'개수 + '!'로 구성 --> nav51!
 - [출력문장] 'http://naver.com' 의 비밀번호는 nav51!입니다.



```python
url = "http://naver.com"

# 규칙 1
str1 = url[7:]

# 규칙 2
str2 = str1[:str1.index(".")]

# 규칙 3
str3 = str2[0:3]
str4 = (str)(len(str2))
str5 = (str)(str2.count('e'))

# 출력문장
print(url + "의 비밀번호는 " + str3 + str4 + str5 + "!" + " 입니다.")

# 답
url_str = url[7:]
url_str1 = url.replace(url_str, ".")

url_str = url_str[:url_str.index(".")]
url_str1 = url_str1[:url_str1.index(".")]

url_str = url_str[:3] + str(len(url_str)) + str(url_str.count("e")) + "!"
print(url + "의 비밀번호는 " + url_str + " 입니다.")
```

# list comprehension
- 리스트를 쉽게, 짧게 한 줄로 만들 수 있는 파이썬의 문법
- 기존 리스트형을 사용하여 간단하게 새로운 리스트를 만드는 기법
- [(변수를 활용한 값) for (사용할 변수 이름) in (순회할 수 있는 값)]

```python
result = []

for i in range(10):
    result.append(i)
print(result)

# Same way to use list comprehension
# result = [i for i in range(10)]
# print(result)
```

#다음과 같이 출력되도록 for문으로 작성하고 List Comprehension으로 나타내시오.
- a = [1, 2, 3, 4, 5, 6, 7, 8, 9]를 이용한다.
- 짝수만 빼서 2배한 값으로 리스트를 만들어라
- 출력리스트 : [4, 8, 12, 16]

```python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# result = []
# for i in a:
#     if i % 2 == 0:
#         result.append(i * 2)
# print(result)

result = []
result = [i * 2 for i in a if i % 2 == 0]
print(result)
```

```python
a = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# result = []
# for i in a:
#     if i % 2 == 0:
#         result.append(i * 2)
#     else:
#         result.append(-1)
# print(result)

result = []
result = [i * 2  if (i % 2 == 0) else -1 for i in a]
print(result)
```

성적 리스트 = [99, 23, 80, 79, 65, 31, 0, 24, 54, 35, 100, 72]를
입력받아 평균을 출력하는 print_score함수를 정의하라

```python
score_list = [99, 23, 80, 79, 65, 31, 0, 24, 54, 35, 100, 72]

def print_score(score_list):
    sum = 0
    for i in range(len(score_list)):
        sum += score_list[i]
    print(int(sum / len(score_list)))

print_score(score_list)

# 선생님 코드
# def print_score(score_list):
#     print(sum(score_list) / len(score_list))

# print_score([99, 23, 80, 79, 65, 31, 0, 24, 54, 35, 100, 72])
```

```python
# 헤더 저장하기
# 헤더 : 각 열의 의미, 제목 : title

from google.colab import files
import csv

data_read = open("/content/drive/MyDrive/2021 빅데이터분석/30604 김강민/Data/seoul.csv", 'r', encoding="cp949")

data = csv.reader(data_read, delimiter=",")
header = next(data)

tem = -100 # 최고기온
date = "" # 최고기온 날짜

for row in data:
    if row[-1] == "":
        row[-1] = -100
    row[-1] = float(row[-1])

if tem < row[-1]:
    tem = row[-1]
    date = row[0]

data_read.close()

print("서울의 최고기온 날짜 : {}, 온도 : {}".format(date, tem))
```
