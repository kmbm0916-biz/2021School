13. 사용자로 입력받은 단어가 아래 fruit 리스트에 포함되어 있는지를 확인하라. 포함되었다면 "집에 있음"을 아닐경우 "구입하세요"를 출력하라.
- fruit["사과", "딸기", "체리"]
- 출력예시 : 과일을 입력하세요 : 키위
- 키위를 구입하세요.

```python
data_fruit = ["사과", "딸기", "체리"]
print("과일을 입력하세요 : ", end="")
input_data = input()

if input_data in data_fruit:
    print("집에 있음")
else:
    print("{}를 구입하세요.".format(input_data))
```

14. 아래와 같이 fruit 딕셔너리가 정의되어 있다. 사용자가 입력한 값이 딕셔너리 키(key)값에 포함되었다면 "정답입니다"를 아닐경우 "오답입니다"를 출력하라.
- fruit : {"봄" : "딸기", "여름" : "토마토", "가을" : "사과"}
- 출력예시 : 제가 좋아 하는 계절은 : 봄
- 정답입니다.

```python
fruit = {"봄" : "딸기", "여름" : "토마토", "가을" : "사과"}
input_data = input("제가 좋아하는 계절은 : ")
if input_data in fruit:
    print("정답입니다.")
else:
    print("오답입니다.")
```

15. 아래와 같이 fruit 딕셔너리가 정의되어 있다. 사용자가 입력한 값이 딕셔너리 값(value)에 포함되었다면 "정답입니다"를 아닐 경우 "오답입니다" 출력하라.
- fruit = {"봄" : "딸기", "여름" : "토마토", "가을" : "사과"}
- 출력예시 : 좋아하는 과일은? 한라봉
- 오답입니다.

```python
fruit = {"봄" : "딸기", "여름" : "토마토", "가을" : "사과"}
input_data = input("좋아하는 과일은? ")
if input_data in fruit.values():
    print("정답입니다.")
else:
    print("오답입니다.")
```

16. 휴대폰 번호 앞자리에 따라 통신사는 아래와 같이 구분된다. 사용자로부터 휴대 전화 번호를 입력 받고, 통신사를 출력하는 프로그램을 작성하라.

```python
input_data = input("휴대전화 번호 입력 : ")
num = input_data.split("-")[0]

if num == "011":
    print("당신은 {} 사용자입니다.".format("SKT"))
elif num == "016":
    print("당신은 {} 사용자입니다.".format("KT"))
elif num == "019":
    print("당신은 {} 사용자입니다.".format("LGU"))
else:
    print("당신은 {} 사용자입니다.".format("알수없음"))
```
