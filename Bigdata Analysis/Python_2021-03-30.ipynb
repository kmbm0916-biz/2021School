22. 리스트에 5일간의 저가, 고가 정보가 저장되어 있다. 
- low_prices = [100, 200, 400, 800, 1000]
- high_prices = [150, 300, 430, 880, 1000]
- 고가와 저가의 차를 변동폭이라고 정의할 때, low, high 두 개의 리스트를 사용해서 5일간의 변동폭을 volatility 리스트에 저장하고, volatility 리스트의 요소들을 출력하라.
- 출력예시 : 고가와 저가의 차의 변동폭은 [50, 100, 30, 80, 0]

```python
low_prices = [100, 200, 400, 800, 1000]
high_prices = [150, 300, 430, 880, 1000]

volatility = []

for i in range(0, len(low_prices)):
    volatility.append(high_prices[i] - low_prices[i])

print("고가와 저가의 차의 변동폭은")
print(volatility)

# 2안(김시현)
# sub = [0, 0, 0, 0, 0]
# for i in range(0, 5):
#     sub[i] = (int(high_prices[i]) - int(low_prices[i]))
# print("고가와 저가의 변동폭", sub)
```

**11-2 while문**

```python
i = 1
while (i < 4):
    print("{}번 학생".format(i))
    i += 1
print("end")
```

- 1부터 100까지의 합을 while문을 이용하여 구하고 출력하라
- 출력형태 : 1부터 100까지의 합은 5050입니다.

```python
i = 1
sum = 0
while (i < 101):
     sum += i
     i += 1
print("1부터 100까지의 합은 {} 입니다.".format(sum))
```
