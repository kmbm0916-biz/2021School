```
# a 리스트에서 중복을 제거하라
a = [1, 1, 1, 2, 2, 3, 3, 3, 4, 4, 5]
b = []
for i in a:
    if i not in b:
        b.append(i)
print(b)
```

```
# 딕셔너리를 선언하라. A : 90, B : 80, C : 70, D : 60, E : 50
dic_a = {"A" : 90, "B" : 80, "C" : 70, "D" : 60, "E" : 50}
print(dic_a)
```
