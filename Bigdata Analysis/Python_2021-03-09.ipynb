**3-1 단순한 연산자**

```
print(1 + 1)
# 더하기
print(1 - 1)
# 빼기
print(2 * 2)
# 곱하기
print(4 / 2)
# 나누기
print(4 % 3)
# 나머지
```

```
a = 10
a += 2 # a = a + 2
print(a)
a *= 2 # a = a * 2
print(a)
a /= 2 # a = a / 2
print(a) # -> 캐스트(형변환) 소수형
a -= 2 # a = a - 2
print(a) # -> 캐스트(형변환) 소수형
a %= 2 # a = a % 2
print(a) # -> 캐스트(형변환) 소수형
```

**3-2 함수를 이용한 연산**

```
# 절댓값 함수
print(abs(-9))
print(abs(9))

# 최댓값 함수
print(max(3, 10, 59, 2, -9))

# 최솟값 함수
print(min(5, 2, 6, -2, 68))

# 랜덤 함수
from random import *
print(random()) # 무작위성, 0.XXXXXXXXXXXXXXXXX로 나온다, 0.0 ~ 1.0 미만의 임의의 값이 만들어진다.
print(random() * 10) # 0 ~ 10 미만의 임의의 값이 만들어진다.
print((int)(random() * 10)) # 0 ~ 10 미만의 임의의 정수값이 만들어진다.
print((int)(random() * 10 + 1)) # 1 ~ 10 이하의 임의의 정수값이 만들어진다.
print((int)(random() * 50 + 1)) # 1 ~ 50 이하의 임의의 정수값이 만들어진다.
print(randrange(1, 51)) # 1 ~ 50 이하의 임의의 정수값이 만들어진다. randrange 함수 이용.
print(randint(1, 50)) # 1 ~ 50 이하의 임의의 정수값이 만들어진다. randint 함수 이용.
```

**3-4 슬라이싱**

```
jumin_num = "030101-3456789"
print("성별 : " + jumin_num[7]) # 8번째 숫자인 성별 출력
print("생년 : " + jumin_num[0:2]) # 생년(03년) 출력
print("생월 : " + jumin_num[2:4]) # --> N:M = N부터 M - 1 까지
print("생일 : " + jumin_num[4:6])
print("생년월일 : " + jumin_num[:6]) # --> == [0:6] 
print("주민등록번호 뒷자리 : " + jumin_num[7:14])
print("주민등록번호 : " + jumin_num[0:6] + jumin_num[7:14])

print(jumin_num[-1]) # 뒤에서부터 첫번째 값
print(jumin_num[-2]) # 뒤에서부터 두번째 값
```

**3-5 문자열 처리 함수**

```
python = 'Python is Amazing'
print(python.lower()) # 문자열 내부 대문자 소문자로 변환
print(python.upper()) # 문자열 내부 소문자 대문자로 변환

print(python[0].isupper()) # 문자열 내부 첫번째 값이 대문자일 경우 True를, 아닐 경우 False를 반환한다.
print(python[0].islower()) # 문자열 내부 첫번째 값이 소문자일 경우 True를, 아닐 경우 False를 반환한다.

print(len(python)) # 문자열의 길이를 출력

print(python.replace('Python', 'Java')) # 문자열 내부에(replace(N, M)) N이 있을 경우 M으로 치환한다.

n = python.index('t')
print(n) # 문자열 내부에(index(N)) N이 있을 경우 N이 처음으로 나오는 인덱스 반환

print(python.find('Python')) # 문자열 내부에(find(N)) N이 있을 경우 N의 위치를 반환(N이 문자열인 경우 가장 첫번째 글자의 위치 반환), 없는 경우 -1

print(python.count('n')) # 문자열 내부에(count(N)) N이 있을 경우 N의 갯수를 반환.

print("나는 {}살 입니다.".format(20)) # 문자열 내부에 format 함수를 통한 값 대입
a = "나는 {}살 입니다."
print(a.format(30)) # 변수 속 문자열 내부에 format 함수 적용
print("나는 {}색과 {}색을 좋아합니다.".format("청록", "빨간")) # format 함수를 통한 값 대입
print("나는 {1}색과 {0}색을 좋아합니다.".format("청록", "빨강")) # 인덱스 위치에 따른 format 함수의 위치지정
print("나는 {age}살이고 {color}색을 좋아합니다.".format(age = 20, color = "빨강")) # 인덱스를 변수 형태로 선언해 위치지정

age = 20
color = "빨강"
print(f"나는 {age}살이고 {color}색을 좋아한다.") # 변수에 선언된 값을 문자열 내부에 대입(format의 약자인 f 사용)
```

**3-6 과제**
> 사이트별로 비밀번호를 만들어 주는 프로그램을 작성하라.
 - "http://"부분은 제외 --> naver.com
 - 처음 만나는 점(.)이후 부분은 제외 --> naver
 - 남은 글자 중 처음 세자리 + 글자 개수 + 글자 내 'e'개수 + '!'로 구성 --> nav51!
 - [출력문장] 'http://naver.com' 의 비밀번호는 nav51!입니다.



```
url = "http://naver.com"

# 규칙 1
str1 = url[7:]

# 규칙 2
str2 = str1[:str1.index(".")]

# 규칙 3
str3 = str2[0:3]
str4 = (str)(len(str2))
str5 = (str)(str2.count('e'))

# 출력문장
print(url + "의 비밀번호는 " + str3 + str4 + str5 + "!" + " 입니다.")

# 답
url_str = url[7:]
url_str1 = url.replace(url_str, ".")

url_str = url_str[:url_str.index(".")]
url_str1 = url_str1[:url_str1.index(".")]

url_str = url_str[:3] + str(len(url_str)) + str(url_str.count("e")) + "!"
print(url + "의 비밀번호는 " + url_str + " 입니다.")
```
