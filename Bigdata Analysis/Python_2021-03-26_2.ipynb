# 10 튜플
**10-2 튜플 과제**

```python
def print_arithmetic_operation(a, b):
    print("{} + {} = {}".format(a, b, a+b))
    print("{} - {} = {}".format(a, b, a-b))
    print("{} x {} = {}".format(a, b, a*b))
    print("{} / {} = {}".format(a, b, a/b))

input_data1 = input("숫자를 입력하세요 : ")
input_data2 = input("숫자를 입력하세요 : ")
print_arithmetic_operation(int(input_data1), int(input_data2))
```
