12. 사용자로부터 하나의 숫자를 입력 받고 짝수/홀수를 판별하라.

```python
print("숫자를 입력하세요 : ", end="")
input_data = input()
if int(input_data) == 0:
    print("0")
elif int(input_data) % 2 == 0:
    print("짝수")
else:
    print("홀수")
# if int(input_data) == 0:
#     print("0")
# elif int(input_data) % 2:
#     print("홀수")
# else:
#     print("짝수")
```
