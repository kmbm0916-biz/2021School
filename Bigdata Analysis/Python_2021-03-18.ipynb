# 7 조건문과 반복문
**7-1 조건문, 반복문 복습**

```
weather = "미세먼지"

if weather == "비":
    print("Bring your umbrella")
elif weather == "미세먼지":
    print("Bring your mask")
else:
    print("Bring nothing!")
```

```
input_climate = input("오늘 날씨 어때요?\n")
if input_climate == "비":
    print("Bring your umbrella")
elif input_climate == "미세먼지":
    print("Bring your mask")
else:
    print("Bring nothing!")
```

```
input_temp = input("오늘 온도는?\n")
if int(input_temp) >= 30:
    print("Too hot!")
elif int(input_temp) >= 20:
    print("Normal temp")
elif int(input_temp) >= 10:
    print("It's cold outside")
else:
    print("Too cold!")
```

```
for waiting_no in [0, 1, 2, 3, 4]:
    print("대기번호 : {}".format(waiting_no))
# for 내부에 선언된 변수는 for문 내부에서만 작동한다. for문 외부에서 변수를 사용할 수 없다.

for waiting_no in range(0, 5):
    print("대기번호 : {}".format(waiting_no))
# range()를 활용한 범위 지정

st = ['aaa', 'bbb', 'ccc']
for waiting_no in st:
    print("대기번호 : {}".format(waiting_no))
# 리스트 내부의 문자도 반환 가능
```

**7-2 간단한 while문**

```
customer = "Tom"
index = 5

while index >= 1:
    print("{0}, 커피가 준비 되었습니다. {1}번 남았습니다.".format(customer, index))
    index -= 1
    if index == 0:
        print("커피가 폐기 처분 되었습니다.")
```

```
person = "Dudi"
person_input = ""

while person != person_input:
    print("{}, 커피가 준비되었습니다.".format(customer))
    person_input = input("Name : ")
    if(person_input != person):
        print("{}, 번호를 다시 확인해 주십시오.".format(person_input))
        break
    elif(person_input == person):
        print("{}, 커피를 받아가십시오.".format(person))
```

**7-3 Continue, break**

```
absent = [2, 5]
no_book = [7]

for student in range(1, 11):
    print(student)
    if student in absent:
        continue
    elif student in no_book:
        print("{}는 교무실로 따라와".format(student))
        break
    print("{}, 조용히 자습해라".format(student))
```

**7-4 과제**

- 출석번호가 1, 2, 3, 4 앞에 100을 더한 값으로 만들기
- 101, 102, 103, 104

```
student = [1, 2, 3, 4]
print(student)

for student in range(1, 5):
    print(student + 100, end = " ")
```
